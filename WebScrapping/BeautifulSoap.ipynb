```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
import re

page = requests.get("https://www.imdb.com/search/title/?title_type=documentary")
soup = BeautifulSoup(page.content, 'html.parser')
conteudo = soup.find_all(class_="lister-item-content")

pos = [cd.find(class_="lister-item-index unbold text-primary").get_text() for cd in conteudo]
name = [cd.find('a').get_text() for cd in conteudo]
year = [int(re.findall('\d+',cd.find(class_="lister-item-year text-muted unbold").get_text())[0]) for cd in conteudo]
desc = [cd.find_all(class_="text-muted")[2].get_text() for cd in conteudo]
duration = [cd.find(class_="runtime").text for cd in conteudo]
genre = [cd.find(class_="genre").text for cd in conteudo]

df = pd.DataFrame({
    "Pos": pos,
    "Name": name,
    "Year": year,
    "Description":desc,
    "Duration": duration,
    "Genre": genre
})

df['Pos'] = df['Pos'].map(lambda x: x.rstrip('.'))
df
```
